```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('flights_encoded.csv')
```

```python
df = df.drop(['origin_city_name', 'dest_city_name', 'year'], axis = 1)
```

```python
df.head()
```

```python
def categorize(df, column):
    df[column] = df[column].astype('category')
    return df[column]
```

```python
df['scheduled_flight_hour_of_day'] = categorize(df, 'scheduled_flight_hour_of_day')
```

```python
df['day_of_the_week'] = categorize(df, 'day_of_the_week')
```

```python
df['origin_conditions'] = categorize(df, 'origin_conditions')
```

```python
df['dest_conditions'] = categorize(df, 'dest_conditions')
```

```python
df['holiday'] = categorize(df, 'holiday')
```

```python
df['month'] = categorize(df, 'month')
```

```python
y =df['arr_delay']
col_names = ['scheduled_flight_hour_of_day', 'day_of_the_week','origin_conditions','dest_conditions','holiday', 'month']
cat_vars_final = df.select_dtypes(['object','category'])
cat_vars_final = cat_vars_final.loc[:, col_names]

#One Hot Encoder

enc = OneHotEncoder().fit(cat_vars_final)

cat_vars_ohe_final = enc.transform(cat_vars_final).toarray()
cat_vars_ohe_final = pd.DataFrame(cat_vars_ohe_final, index= cat_vars_final.index, 
                      columns=enc.get_feature_names(cat_vars_final.columns.tolist()))
```

```python
cat_vars_ohe_final.head()
```

```python
cat_vars_ohe_final['arr_delay'] = y
```

```python
corr = pd.DataFrame(cat_vars_ohe_final[cat_vars_ohe_final.columns].corr()['arr_delay'][:])
```

```python
corr 
```

```python
ind = list(corr.index.values)
```

```python
cat_vars_ohe_final = cat_vars_ohe_final.drop(ind, axis = 1)
```

```python
cat_vars_ohe_final
```

```python
cat_vars_ohe_final[cat_vars_ohe_final.columns].corr()['arr_delay'][:]
```

```python
df = df.drop(col_names, axis = 1)
```

```python
df
```

```python
cat_vars_ohe_final
```

```python
cat_vars_ohe_final[['distance','origin_wind_speed','origin_visibility','dest_wind_speed','dest_visibility','arr_delay']] = df
```

```python
cat_vars_ohe_final
```

```python
y = cat_vars_ohe_final['arr_delay']
```

```python
X = cat_vars_ohe_final.drop(['arr_delay'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.7, random_state = 5)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler(feature_range=(-1,1)).fit(X_train)
X_train = scaling.transform(X_train)
X_test = scaling.transform(X_test)
```

```python
svr = SVR(kernel = 'linear')
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_2_score = r2_score(y_test, y_pred)
```

```python
print(r_2_score)
```

```python
df1 = pd.read_csv('flights_encoded.csv')
```

```python
df1 = df1.drop(['origin_city_name', 'dest_city_name', 'year'], axis = 1)
```

```python
def categorize(df, column):
    df[column] = df[column].astype('category')
    return df[column]
```

```python
df1['scheduled_flight_hour_of_day'] = categorize(df1, 'scheduled_flight_hour_of_day')
```

```python
df1['day_of_the_week'] = categorize(df1, 'day_of_the_week')
```

```python
df1['origin_conditions'] = categorize(df1, 'origin_conditions')
```

```python
df1['dest_conditions'] = categorize(df1, 'dest_conditions')
```

```python
df1['holiday'] = categorize(df1, 'holiday')
```

```python
df1['month'] = categorize(df1, 'month')
```

```python
y =df1['arr_delay']
col_names = ['scheduled_flight_hour_of_day', 'day_of_the_week','origin_conditions','dest_conditions','holiday', 'month']
cat_vars_final = df1.select_dtypes(['object','category'])
cat_vars_final = cat_vars_final.loc[:, col_names]

#One Hot Encoder

enc = OneHotEncoder().fit(cat_vars_final)

cat_vars_ohe_final = enc.transform(cat_vars_final).toarray()
cat_vars_ohe_final = pd.DataFrame(cat_vars_ohe_final, index= cat_vars_final.index, 
                      columns=enc.get_feature_names(cat_vars_final.columns.tolist()))
```

```python
df1 = df1.drop(col_names, axis = 1)
```

```python
df1
```

```python
cat_vars_ohe_final[['distance','origin_wind_speed','origin_visibility','dest_wind_speed','dest_visibility','arr_delay']] = df1
```

```python
cat_vars_ohe_final = cat_vars_ohe_final.drop(['scheduled_flight_hour_of_day_1', 'day_of_the_week_1', 'origin_conditions_1','dest_conditions_1', 'holiday_1', 'month_1'], axis = 1)
```

```python
X = cat_vars_ohe_final.drop(['arr_delay'], axis = 1)
```

```python
y = cat_vars_ohe_final['arr_delay']
```

```python
import xgboost as xgb
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.7, random_state = 5)
```

```python
from sklearn.model_selection import GridSearchCV
xg = xgb.XGBRegressor(n_estimators = 100)
```

```python
param_grid={"learning_rate": (0.05, 0.10, 0.15),
                        "max_depth": [ 3, 4, 5, 6, 8],
                        "min_child_weight": [ 1, 3, 5, 7],
                        "gamma":[ 0.0, 0.1, 0.2],
                        "colsample_bytree":[ 0.3, 0.4]}

# define the grid search
XG_reg= GridSearchCV(xg, param_grid,cv=5)

#fit the grid search
XG_reg.fit(X_train,y_train)

# best estimator
print(XG_reg.best_estimator_)
```

```python
# best model
best_model = XG_reg.best_estimator_
best_model.fit(X_train,y_train)
```

```python
y_pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_2_score = r2_score(y_test, y_pred)
print(r_2_score)
```
