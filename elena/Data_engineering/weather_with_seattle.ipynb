```python
import pandas as pd
```

```python
weather = pd.read_csv('test_weather.csv')
```

```python
seattle = pd.read_csv('seattle_jan_2020.csv')
```

```python
weather['Name'].unique()
```

```python
weather = weather[weather['Name'] != 'Phoenix, AZ, United States']
```

```python
weather['Name'].unique()
```

```python
weather = pd.concat([weather, seattle],ignore_index=True)
```

```python
weather['Name'].unique()
```

```python
my_lambda = lambda x: ''.join(x.split(',')[0]).lower()
weather['Name'] = weather['Name'].apply(my_lambda)
```

```python
weather['Name'].replace('fort worth ave', 'dallas/fort worth', inplace=True)
```

```python
weather['Name'].unique()
```

```python
weather['Date time'] = pd.to_datetime(weather['Date time'])
weather['Date time'] = weather['Date time'].dt.strftime('%Y-%m-%d')
```

```python
weather_wvc = weather[['Name', 'Date time', 'Wind Speed', 'Visibility', 'Conditions']]
```

### Seattle test_data

```python
sea_test = pd.read_csv('seattle_test_flights.csv')
```

```python
sea_test.head()
```

```python
sea_test['origin_city_name'] = sea_test['origin_city_name'].apply(my_lambda)
sea_test['dest_city_name'] = sea_test['dest_city_name'].apply(my_lambda)
```

```python
### Adding columns based on origin city name
test_seattle_weather = pd.merge(sea_test, weather_wvc, how = 'left', left_on=['origin_city_name', 'fl_date'], right_on=['Name', 'Date time']).rename(columns={"Wind Speed": "origin_wind_speed",
                                                                                                                                            "Visibility": "origin_visibility",
                                                                                                                                            "Conditions": "origin_conditions"})
```

```python
### Adding columns based on dest city name
test_seattle_weather = pd.merge(sea_test, weather_wvc, how = 'left', left_on=['dest_city_name', 'fl_date'], right_on=['Name', 'Date time']).rename(columns={"Wind Speed": "dest_wind_speed",
                                                                                                                                            "Visibility": "dest_visibility",
                                                                                                                                            "Conditions": "dest_conditions"})
```

```python
test_seattle_weather.tail()
```

```python
test_seattle_weather = test_seattle_weather.drop(['Name', 'Date time'], axis = 1)
```

```python
test_seattle_weather['day_of_the_week'] = pd.to_datetime(test_seattle_weather['fl_date']).dt.day_name()
```

```python
test_seattle_weather['fl_date'] = pd.to_datetime(test_seattle_weather['fl_date'])
```

```python
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
cal = calendar()
```

```python
holidays = cal.holidays(start='2020-01-01', end='2020-01-31')
test_seattle_weather['holiday'] = test_seattle_weather['fl_date'].isin(holidays)
```
