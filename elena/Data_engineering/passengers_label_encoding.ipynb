```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('flights_2019_pass.csv')
```

```python
flights.head()
```

```python
column_names = list(flights.columns.values)
```

```python
print(column_names)
```

```python
columns = ['origin_airport_id', 'dest_airport_id', 'origin_city_name','dest_city_name', 'distance', 'day_of_the_week', 'origin_wind_speed', 'origin_visibility', 'origin_conditions', 'dest_wind_speed', 'dest_visibility', 'dest_conditions', 'holiday','arr_delay', 'passengers', 'month']
```

```python
df_flights = flights[columns]
```

```python
df_flights.head()
```

```python
df_flights['arr_delay'].isnull().value_counts()
```

```python
df_flights = df_flights.dropna()
```

```python
df_flights['arr_delay'].isnull().value_counts()
```

```python
def label_encoding(df, column):
    df[column] = df[column].astype('category')
    df[column] = df[column].cat.codes.astype('category')
    return df[column]
```

```python
df_flights['origin_city_name'] = label_encoding(df_flights, 'origin_city_name')
```

```python
df_flights['dest_city_name'] = label_encoding(df_flights, 'dest_city_name')
```

```python
df_flights['day_of_the_week'] = label_encoding(df_flights, 'day_of_the_week')
```

```python
df_flights['origin_conditions'] = label_encoding(df_flights, 'origin_conditions')
```

```python
df_flights['dest_conditions'] = label_encoding(df_flights, 'dest_conditions')
```

```python
df_flights['holiday'] = label_encoding(df_flights, 'holiday')
```

```python
df_flights.to_csv('delta_encoded.csv', sep=",", index = False)
```
