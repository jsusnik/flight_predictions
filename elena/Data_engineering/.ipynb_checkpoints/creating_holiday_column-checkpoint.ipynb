```python
import pandas as pd
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
```

```python
flights_with_weather = pd.read_csv('flights_with_weather.csv')
```

```python
test_flights_with_weather = pd.read_csv('test_flights_with_weather.csv')
```

```python
flights_with_weather.head()
```

```python
flights_with_weather['fl_date'] = pd.to_datetime(flights_with_weather['fl_date'])
```

```python
cal = calendar()
```

### Creating holiday column for flights_with_weather df

```python
holidays = cal.holidays(start='2018-01-01', end='2019-12-31')
```

```python
flights_with_weather['holiday'] = flights_with_weather['fl_date'].isin(holidays)
```

```python
flights_with_weather.head()
```

```python
flights_with_weather.to_csv('flights_weather_holiday.csv', sep=",", index = False)
```

### Creating holiday column for test_flights_with_weather df

```python
holidays = cal.holidays(start='2020-01-01', end='2020-01-31')
```

```python
test_flights_with_weather['holiday'] = test_flights_with_weather['fl_date'].isin(holidays)
```

```python
test_flights_with_weather.head()
```

```python
test_flights_with_weather.to_csv('test_flights_weather_holiday.csv', sep=",", index = False)
```
