```python
import pandas as pd
```

```python
df1 = pd.read_csv('flights_jan_dec_18.csv')
```

```python
df2 = pd.read_csv('flights_jan_dec_19.csv')
```

```python
flights = pd.concat([df1, df2], ignore_index=True)
```

```python
flights = flights.drop(['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay'], axis = 1)
```

### Creating a column with day of the week names

```python
flights['day_of_the_week'] = pd.to_datetime(flights['fl_date']).dt.day_name()
```

### Changing city names format so it will correspond to a weather format

```python
my_lambda = lambda x: ''.join(x.split(',')[0]).lower()
```

```python
flights['origin_city_name'] = flights['origin_city_name'].apply(my_lambda)
```

```python
flights['dest_city_name'] = flights['dest_city_name'].apply(my_lambda)
```

### Creating weather columns

```python
weather = pd.read_csv('weather.csv')
```

```python
weather_wvc = weather[['Name', 'Date time', 'Wind Speed', 'Visibility', 'Conditions']]
```

```python
### Adding columns based on origin city name
flights = pd.merge(flights, weather_wvc, how = 'left', left_on=['origin_city_name', 'fl_date'], right_on=['Name', 'Date time']).rename(columns={"Wind Speed": "origin_wind_speed",
                                                                                                                                            "Visibility": "origin_visibility",
                                                                                                                                            "Conditions": "origin_conditions"})
```

```python
### Adding columns based on dest city name
flights = pd.merge(flights, weather_wvc, how = 'left', left_on=['dest_city_name', 'fl_date'], right_on=['Name', 'Date time']).rename(columns={"Wind Speed": "dest_wind_speed",
                                                                                                                                            "Visibility": "dest_visibility",
                                                                                                                                            "Conditions": "dest_conditions"})
```

```python
flights = flights.drop(['Name_x', 'Date time_x', 'Name_y', 'Date time_y'], axis = 1)
```
